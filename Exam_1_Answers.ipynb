Problem 5
With the command > new_file, I can rewrite in a different file the content of an existing document. In fact, I can use it together with other commands as head or tail to choose the lines of the files we want to rewrite. In this case I want to create a document that contains only the data lines and exlude the header line. To achieve that, I have to type tail -7215 to select all the data lines and then use > Data_Only.csv to rewrite them.      

```bash
tail -7215 Locus_Germplasm_Phenotype_20130122.txt > Data_Only.csv
wc -l Data_Only.csv
head -5 Data_Only.csv
```

The commands wc -l and head -5 prove that the code has worked as expected and that the data lines have been rewritten in a new file.

Problem 7
First of all, I used the same command (except for the -o) to select the lines that have a phenotype associated with the word root in the file. Then I used a pipe and the comand grep -E -in '\AT[0-9]G[0-9]{5}' to obtain the Locus codes included in those lines.

```bash
grep -n '\root' Locus_Germplasm_Phenotype_20130122.txt | grep -E -in '\AT[0-9]G[0-9]{5}'
```

As we can see, the output creates a new list starting with the number 1 that corresponds to the first Locus code with a phenotype where the word root is included. However, it also displays a lot information we don't want to know right now. To eliminate it, I added a -o in the last grep command so now it will only display the Locus codes (the list will start with the number 1 again). To rewrite all of them in a new file I typed > Root-associated-Loci.txt and cat Root-associated-Loci.txt to check if it worked properly.      

```bash
grep -n '\root' Locus_Germplasm_Phenotype_20130122.txt | grep -E -ino '\AT[0-9]G[0-9]{5}' > Root-associated-Loci.txt
cat Root-associated-Loci.txt
```
