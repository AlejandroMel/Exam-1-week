Problem 1
To prove that the lanes in Germplasm and LocusGene are in the same sequence, I will open both files and then with the function DictReader ignore the header line. After that i will use the function for to create a loop and create a bunch of lists that will place the locus codes of both files together. Finally, I will use the function if to create a conditional that will assess if the locus codes in both files are the same. 

```python
import csv
csvfile = open("LocusGene.tsv", "r")
locusgene_data = csv.DictReader(csvfile, delimiter="\t", quotechar = '"')
csvvfile = open("Germplasm.tsv", "r") 
germplasm_data = csv.DictReader(csvvfile, delimiter="\t", quotechar = '"')             
    
for row in locusgene_data:
    for row in germplasm_data:
        a = [row["Locus"]]
        b = [row["Locus"]]
        print(a+b)    
        if a == b:
            print ("Same sequence")
        else:
            print ("No match!")
            
csvfile.close()
csvvfile.close()
```

Problem 3
To fill the database, I will use the library pymysql.cursors. First of all, I will connect to the database and open the files to obtein the data. Then I will use the function try folloed by with and for to create a loop and be able to insert all of the data in both tables. 

```python
import pymysql.cursors
import csv
locusfile = open("LocusGene.tsv", "r")
locusgene_data = csv.DictReader(locusfile, delimiter="\t", quotechar = '"')
germfile = open("Germplasm.tsv", "r") 
germplasm_data = csv.DictReader(germfile, delimiter="\t", quotechar = '"')
connection = pymysql.connect(host='localhost',
                             user='root',
                             password='root',
                             db='germplasm',
                             charset='utf8mb4',  
                             cursorclass=pymysql.cursors.DictCursor)
try:
    with connection.cursor() as cursor:
        for row in locusgene_data:
            sql = """INSERT INTO LocusGene (agi_locus_code, gene, protein_length)
            VALUES ('{}', '{}', '{}')""".format(row["Locus"], row["Gene"], row["ProteinLength"])
            cursor.execute(sql)
        for row in germplasm_data:
            sql = """INSERT INTO Germplasm (agi_locus_code, germplasm, phenotype, pubmed)
            VALUES ('{}', '{}', '{}', '{}')""".format(row["Locus"], row["germplasm"], row["phenotype"], row["pubmed"])
            cursor.execute(sql)
    connection.commit()
finally:
    print("")

```

To complete reports 1 and 2 I will use the library pymysql.cursors. For report 1 I will connect to the database and then join both tables indicating that the locus codes are the same. For report 2 I will select both tables and the functions AND and OR to select the genes SKOR and MAA3 in particular.   

```python
#report 1
import pymysql.cursors
connection = pymysql.connect(host='localhost',
                             user='root',
                             password='root',
                             db='germplasm',
                             charset='utf8mb4',
                             cursorclass=pymysql.cursors.DictCursor)
try:
    with connection.cursor() as cursor:
        sql = """SELECT g.*, l.* FROM Germplasm g INNER JOIN LocusGene l ON
        g.agi_locus_code = l.agi_locus_code"""
        cursor.execute(sql)
        results = cursor.fetchall()
        print(results)
finally:
    print("")
    connection.close()
```

```python
#report 2
connection = pymysql.connect(host='localhost',
                             user='root',
                             password='root',
                             db='germplasm',
                             charset='utf8mb4',
                             cursorclass=pymysql.cursors.DictCursor)
connection.autocommit(True)

try:
    with connection.cursor() as cursor:
        sql = """SELECT * FROM Germplasm, LocusGene WHERE 
        LocusGene.agi_locus_code = Germplasm.agi_locus_code AND (LocusGene.gene = 'SKOR' OR LocusGene.gene = 'MAA3')"""
        cursor.execute(sql)
        results = cursor.fetchall() 
        print(results)

finally:
    print("")
connection.close()
```
